```python
import os
import re

# Make sure the output root exists
os.makedirs(output_root, exist_ok=True)

sadhu_words = [

    "করিল", "খাইল", "দেখিল", "শুনিল", "লিখিল", "বলিল",
    "বসিল", "দাঁড়াইল", "হাসিল", "যাইল", "আসিল", "খেলিল",
    "চিন্তিল", "পড়িল", "চলিল", "গাহিল",


    "করিবে", "যাইবে", "খাইবে", "দেখিবে", "শুনিবে", "লিখিবে",
    "বলিবে", "বসিবে", "দাঁড়াইবে", "হাসিবে", "আসিবে", "খেলিবে",
    "চিন্তিবে", "পড়িবে", "চলিবে", "গাহিবে", "পাইবে",


    "খাইতেছিল", "করিতেছিল", "যাইতেছিল", "দেখিতেছিল", "শুনিতেছিল",
    "লিখিতেছিল", "বলিতেছিল", "বসিতেছিল", "দাঁড়াইতেছিল", "হাসিতেছিল",
    "চলিতেছিল", "পড়িতেছিল", "গাহিতেছিল", "খেলিতেছিল",
    "চিন্তা করিতেছিল", "আসিতেছিল", "নিতে ছিল", "দিতে ছিল"
]
```

```python
def is_sadhu(text, sadhu_words, threshold=50):
    count = 0
    for word in sadhu_words:
        # Use regex to match full word only
        matches = re.findall(rf'\b{re.escape(word)}\b', text)
        count += len(matches)
    return count > threshold
```

```python
for cat in ['essaysH', 'novelsH', 'poemsH', 'storiesH']:
    input_folder = os.path.join(root_input, cat)
    sadhu_output_folder = os.path.join(output_root, f'{cat}_sadhu')
    cholito_output_folder = os.path.join(output_root, f'{cat}_cholito')

    os.makedirs(sadhu_output_folder, exist_ok=True)
    os.makedirs(cholito_output_folder, exist_ok=True)

    for root_dir, dirs, files_list in os.walk(input_folder):
        for filename in files_list:
            file_path = os.path.join(root_dir, filename)
            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    text = f.read()
            except Exception as e:
                print(f"❌ Could not read file: {file_path} — {e}")
                continue

            if is_sadhu(text, sadhu_words):
                out_folder = sadhu_output_folder
                status = "SADHU"
            else:
                out_folder = cholito_output_folder
                status = "CHOLITO"

            out_path = os.path.join(out_folder, filename)
            with open(out_path, 'w', encoding='utf-8') as f_out:
                f_out.write(text)

            print(f"✅ {filename} → {status} → {out_folder}")
```
