```python
import pandas as pd
import unicodedata

# Load Excel files with UTF-8 encoding
file1 = r'D:\Users\Jim\Downloads\Top_Authors.xlsx'
file2 = r'D:\Users\Jim\Downloads\top_authors2.xlsx'

# Read first column only
df1 = pd.read_excel(file1, usecols=[0])
df2 = pd.read_excel(file2, usecols=[0])

# Rename columns
df1.columns = ['Author']
df2.columns = ['Author']

# Function to clean and normalize Bangla text
def normalize_bangla(text):
    if pd.isna(text):
        return ''
    text = str(text).strip()
    text = unicodedata.normalize('NFC', text)  # Normalize Unicode
    return text

# Apply normalization
df1['Author'] = df1['Author'].apply(normalize_bangla)
df2['Author'] = df2['Author'].apply(normalize_bangla)

# Drop duplicates
unique_authors_1 = set(df1['Author'].dropna().unique())
unique_authors_2 = set(df2['Author'].dropna().unique())

# Get unique authors in file2 not in file1
only_in_file2 = sorted(unique_authors_2 - unique_authors_1)

# Print result
print("✅ Unique Bangla authors in top_authors2.xlsx but not in Top_Authors.xlsx:")
for author in only_in_file2:
    print(author)
```
